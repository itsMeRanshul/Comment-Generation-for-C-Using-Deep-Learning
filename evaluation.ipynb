```python
from transformers import RobertaTokenizer, T5ForConditionalGeneration
import torch

model_path = "/content/codet5-comment-gen-final/codet5-comment-gen-final"

# Load tokenizer and model from your fine-tuned directory
tokenizer = RobertaTokenizer.from_pretrained(model_path)
model = T5ForConditionalGeneration.from_pretrained(model_path)
model.eval()
```

```python
import json

with open("/content/structured_dataset_with_ast1.json") as f:
    filtered_data = json.load(f)
```

```python
import random
import evaluate
from tqdm import tqdm
# Load BLEU and ROUGE scorers
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")

sample_data = random.sample(filtered_data, 4000)

# Select a subset for quick evaluation
# sample_data = filtered_data[:100]  # Or use entire filtered_data if you want full evaluation

predictions = []
references = []

model.eval()

for item in tqdm(sample_data):
    input_ids = tokenizer(item["combined_input"], return_tensors="pt", truncation=True, padding=True).input_ids.to(model.device)
    output = model.generate(input_ids, max_length=64)
    decoded_pred = tokenizer.decode(output[0], skip_special_tokens=True)

    predictions.append(decoded_pred)
    references.append(item["comment"])
```

```python
# BLEU requires list of references (as list of list of tokens)
bleu_score = bleu.compute(predictions=predictions, references=[[ref] for ref in references])
rouge_score = rouge.compute(predictions=predictions, references=references)

print("✅ Evaluation Results:")
print("BLEU:", bleu_score["bleu"])
print("ROUGE-L:", rouge_score["rougeL"])
```
